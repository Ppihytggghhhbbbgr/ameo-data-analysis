```python
#frequency distribution of each column

import pandas as pd

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns


frequency_distributions = {}

for column in numerical_columns:
    frequency_distributions[column] = df[column].value_counts().sort_index()

for column, distribution in frequency_distributions.items():
    print(f"Frequency Distribution for {column}:\n{distribution}\n")
```
